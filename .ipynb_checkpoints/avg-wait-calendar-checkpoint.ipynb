```python
import pandas as pd
import numpy as np
from pandas.plotting import table
# import calmap
import matplotlib.pyplot as plt
```

```python
firewhip_may = pd.read_csv("ride-data/firewhip-05-may.csv")
firewhip_may.head()
```

```python
atracoes = ['autopista', 'babyelefante', 'barcopirata', 'betinho4d', 'bigtower', 'carrosselveneziano', 'crazyriver', 'dumdum', 'ferroviadinomagic', 'firewhip', 'pedalinho', 'raskapuska', 'rebulico', 'rodagigante', 'starmountain', 'tchibum', 'tigormountain', 'trenzinhovilaesperanca', 'xicarasmalucas']
nerf = ['spinblast', 'supersoakersplash']

df_atracoes = []

for atracao in atracoes:
    df_jan = pd.read_csv(f"ride-data/{atracao}-01-jan.csv")
    df_feb = pd.read_csv(f"ride-data/{atracao}-02-feb.csv")
    df_mar = pd.read_csv(f"ride-data/{atracao}-03-mar.csv")
    df_apr = pd.read_csv(f"ride-data/{atracao}-04-apr.csv")
    df_may = pd.read_csv(f"ride-data/{atracao}-05-may.csv")
    df_jun = pd.read_csv(f"ride-data/{atracao}-06-jun.csv")
    df_jul = pd.read_csv(f"ride-data/{atracao}-07-jul.csv")
    df_aug = pd.read_csv(f"ride-data/{atracao}-08-aug.csv")
    df_sep = pd.read_csv(f"ride-data/{atracao}-09-sep.csv")
    df_oct = pd.read_csv(f"ride-data/{atracao}-10-oct.csv")
    df_nov = pd.read_csv(f"ride-data/{atracao}-11-nov.csv")
    df_dec = pd.read_csv(f"ride-data/{atracao}-12-dec.csv")

    df_atracao = pd.concat([df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec])

    df_atracoes.append(df_atracao)

for atracao in nerf:
    df_oct = pd.read_csv(f"ride-data/{atracao}-10-oct.csv")
    df_nov = pd.read_csv(f"ride-data/{atracao}-11-nov.csv")
    df_dec = pd.read_csv(f"ride-data/{atracao}-12-dec.csv")
    
    df_atracao = pd.concat([df_oct, df_nov, df_dec])

    df_atracoes.append(df_atracao)

df_geral = pd.concat(df_atracoes)
```

```python
df_geral
```

```python
df_geral['data_hora'] = pd.to_datetime(df_geral['data_hora'])

df_geral['date'] = df_geral['data_hora'].dt.date
```

```python
espera_media = df_geral.groupby('date')['wait_time'].mean().reset_index()
espera_media['wait_time'] = espera_media['wait_time'].round(0)
```

```python
espera_media['date'] = pd.to_datetime(espera_media['date'])

espera_media['year'] = espera_media['date'].dt.year
espera_media['month'] = espera_media['date'].dt.month
espera_media['day'] = espera_media['date'].dt.day

espera_media
```

```python
tabela_horarios = pd.read_excel("bcw-2023.xlsx")
tabela_horarios["date"] = pd.to_datetime(tabela_horarios["date"])
tabela_horarios
```

```python
espera_media = pd.merge(espera_media, tabela_horarios, on="date", how="left")

espera_media
```

```python
# Suponha que vocÃª jÃ¡ tenha o DataFrame "espera_media" com os valores mÃ©dios de fila.

# Configure o Ã­ndice do DataFrame como a coluna de data
espera_media['data'] = pd.to_datetime(espera_media['data'])
espera_media.set_index('data', inplace=True)

# Crie uma matriz vazia para representar o calendÃ¡rio
calendar = np.empty((12, 31))  # 12 meses, 31 dias

# Preencha a matriz com valores de tempo mÃ©dio de fila
for mes in range(1, 13):
    for dia in range(1, 32):
        data = f'2023-{mes:02d}-{dia:02d}'  # Substitua '2023' pelo ano desejado
        if data in espera_media.index:
            calendar[mes - 1][dia - 1] = espera_media.loc[data, 'wait_time']
        else:
            calendar[mes - 1][dia - 1] = np.nan  # Use NaN para dias sem dados

# Reorganize as colunas para comeÃ§ar nas segundas-feiras
calendar = np.roll(calendar, 1, axis=1)

# Crie um mapa de calor para o calendÃ¡rio
plt.figure(figsize=(12, 8))
plt.imshow(calendar, cmap='coolwarm', aspect='auto', interpolation='nearest')
plt.colorbar(label='Tempo MÃ©dio de Fila')
plt.title('Calendar Heatmap do Tempo MÃ©dio de Fila (2023)')

# Configure os rÃ³tulos dos eixos x e y
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
dias_da_semana = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'SÃ¡b', 'Dom']

plt.xticks(np.arange(0, 31), dias_da_semana)
plt.yticks(np.arange(0, 12), meses)

# Rotacione os rÃ³tulos dos eixos y para melhor visualizaÃ§Ã£o
plt.yticks(rotation=0)

# Exiba o mapa de calor do calendÃ¡rio
plt.show()
```
